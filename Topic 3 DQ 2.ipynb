To estimate the probability of falling on heads for a biased coin whose probability of falling on heads is unknown, we can use a statistical approach called the binomial distribution. 
A binomial distribution can be thought of as the probability of success or failure in an experiment or investigation repeated several times. 

```python
#Here, we use the Numpy, random and Matplotlib libraries to simulate the biased coin toss experiment with Python.
import numpy as np
import random
import matplotlib.pyplot as plt

#
#we define the binomial function to landing heads.
#n -> number of tests in an experiment.
#prob_ty -> probability of success / heads.

def event(n,prob_ty):    
    Outcome = np.random.binomial(n,prob_ty)    
    return Outcome
Prob_ty = random.random()

 # Initialize an empty list
    
my_list = [] 
Prob_expt = []
for Trials in range(1, 100):
    

    Outcome = event(Trials,Prob_ty)
    x=Outcome/Trials

    # Add elements to the list
    
    my_list.append(Outcome)  
    
    Prob_expt.append(x)
    

for i in Prob_expt:
    
    print("probability of heads in each Experiment:\n",i)

```

```python
plt.plot(Prob_expt)
plt.xlabel('Number of Heads in each Experiment')
plt.ylabel('Probability of Occurrence',)
plt.show()
```

we see that the probabilities are very different, so it is impossible to determine the probability of heads.
